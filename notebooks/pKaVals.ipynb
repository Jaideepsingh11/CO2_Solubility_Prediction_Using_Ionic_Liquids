```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("../data/DataWithDescriptors.csv")
cation_pka=pd.read_excel("../data/pk_cation.xlsx")
anion_pka=pd.read_excel("../data/pkaanion.xlsx")
print(df.shape)
print(cation_pka.shape)
print(anion_pka.shape)
```

```python
cation_pka=cation_pka.rename(columns={ "CATION": 'CATION NAME' })
```

```python
anion_pka=anion_pka.rename(columns={"Anion":"ANION NAME"})
```

```python
df = pd.merge(df,cation_pka,on="CATION NAME",how="left")
```

```python
df = pd.merge(df,anion_pka,on="ANION NAME",how="left")
```

```python
df["CATION_pKa"].isnull().sum()
```

```python
df["Anion_pKa"].isnull().sum()
```

```python
df["Delta_pKa"] = df["CATION_pKa"] - df["Anion_pKa"]
```

```python
df["Delta_pKa"].describe()
```

```python
df[df["Delta_pKa"] < 2][
    [
        "CATION NAME",
        "ANION NAME",
        "CATION_pKa",
        "Anion_pKa",
        "Delta_pKa"
    ]
]
```

```python
df.describe()
```

```python
df.isnull().sum().sum()
```

```python
miss_val_cols=df.columns[df.isnull().sum()>0]
```

```python
len(miss_val_cols)
```

```python
df=df.drop(columns=miss_val_cols)
```

```python
df=df.drop(columns="Unnamed: 0")
```

```python
df.isnull().sum().sum()
```

```python
const=[]
for col in df.columns:
    if df[col].nunique()==1:
        const.append(col)
```

```python
len(const)
```

```python
df.drop(columns=const,inplace=True)
```

```python
unuseful_cols=['IL NAME', 'CATION NAME', 'ANION NAME', 'Cation_SMILES', 'Anion_SMILES','Cation_Family','Delta_pKa']
```

```python
df_model=df.drop(columns=unuseful_cols)
```

```python
X=df_model.drop(columns=["CO2 capacity (mol CO2/mol IL)"])
Y=df_model["CO2 capacity (mol CO2/mol IL)"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
print(X_train.shape,X_test.shape,y_train.shape, y_test.shape)
```

```python
constant_cols = []

for col in X_train.columns:

    if X_train[col].nunique() == 1:

        constant_cols.append(col)

print("Constant Features:", len(constant_cols))
```

```python
X_train = X_train.drop(
    columns=constant_cols
)

X_test = X_test.drop(
    columns=constant_cols
)
```

```python
from sklearn.feature_selection import VarianceThreshold

selector = VarianceThreshold(threshold=0.001)

selector.fit(X_train)
```

```python
selected_cols=X_train.columns[selector.get_support()]
X_train_low_var = X_train[selected_cols]
X_test_low_var = X_test[selected_cols]
```

```python
print(X_train.shape,X_train_low_var.shape)
```

```python
X_train = X_train_low_var.copy()
X_test = X_test_low_var.copy()
```

```python
corr=X_train.corr().abs()
```

```python
upper = corr.where(np.triu(np.ones(corr.shape),k=1).astype(bool))
```

```python
to_drop = []

for col in upper.columns:

    if any(upper[col] > 0.975):

        to_drop.append(col)
```

```python
len(to_drop)
```

```python
X_train_corr = X_train.drop(
    columns=to_drop
)

X_test_corr = X_test.drop(
    columns=to_drop
)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(
    n_estimators=200,
    random_state=42,
    n_jobs=-1
)

rf.fit(X_train_corr, y_train)
```

```python
importance = pd.DataFrame({
    "Feature": X_train_corr.columns,
    "Importance": rf.feature_importances_
})

importance = importance.sort_values(
    by="Importance",
    ascending=False
)

importance.head(30)
```

```python
rf_top20 = set(importance.head(20)["Feature"])
```

```python
rf_features = importance["Feature"].head(20).tolist()
```

```python
rf_features = importance["Feature"].head(20).tolist()

X_rf_train = X_train_corr[rf_features]
X_rf_test = X_test_corr[rf_features]

print(X_rf_train.shape)
print(X_rf_test.shape)
```

```python
pd.DataFrame(
    rf_features,
    columns=["Descriptor"]
).to_excel("../outputs/tables/RF_Top20_FeaturesWithpKa.xlsx",
    index=False
)
```

```python
X_rf_train.to_csv("../data/X_rf_trainpKa.csv", index=False)
X_rf_test.to_csv("../data/X_rf_testpKa.csv", index=False)

y_train.to_csv("../data/y_trainpKa.csv", index=False)
y_test.to_csv("../data/y_testpKa.csv", index=False)
```
