```python
import pandas as pd
import numpy as np
from rdkit import Chem
import matplotlib.pyplot as plt
import seaborn as sns
from rdkit.Chem import Descriptors
```

```python
df=pd.read_excel("../data/Solubility_data_with_families.xlsx")
```

```python
df.head()
```

```python
m=Chem.MolFromSmiles(df.loc[6,"Cation_SMILES"])
Descriptors.TPSA(m)
```

```python
vals = Descriptors.CalcMolDescriptors(m)
vals['TPSA']
```

```python
mols=[]
for m in df["Cation_SMILES"]:
    mols.append(Chem.MolFromSmiles(m))
```

```python
descrs = [Descriptors.CalcMolDescriptors(mol) for mol in mols]
desc = pd.DataFrame(descrs)
```

```python
desc.sample(7)
```

```python
desc.isnull().sum().sum()
```

```python
desc.columns = ["Cat_" + col for col in desc.columns]
```

```python
desc.head()
```

```python
desc.describe()
```

```python
mols_anion=[]
for m in df["Anion_SMILES"]:
    mols_anion.append(Chem.MolFromSmiles(m))
```

```python
descrs_an = [Descriptors.CalcMolDescriptors(mol) for mol in mols_anion]
desc_an = pd.DataFrame(descrs_an)
```

```python
desc_an.describe()
```

```python
desc_an.isnull().sum().sum()
```

```python
invalid_anions = 0

for smile in df["Anion_SMILES"]:

    mol = Chem.MolFromSmiles(smile)

    if mol is None:
        invalid_anions += 1

print(invalid_anions)
```

```python
desc_an.columns = [
    col.replace("AN_Cat_", "An_")
    for col in desc_an.columns
]
```

```python
df_fin=pd.concat([df,desc,desc_an],axis=1)
```
