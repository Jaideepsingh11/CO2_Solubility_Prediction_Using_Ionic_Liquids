```python
unuseful_cols=['IL NAME', 'CATION NAME', 'ANION NAME', 'Cation_SMILES', 'Anion_SMILES','Cation_Family']
```

```python
df_model=df.drop(columns=unuseful_cols)
df_model.head()
```

```python
X=df_model.drop(columns=["CO2 capacity (mol CO2/mol IL)"])
Y=df_model["CO2 capacity (mol CO2/mol IL)"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
print(X_train.shape,X_test.shape,y_train.shape, y_test.shape)
```

```python
constant_cols = []

for col in X_train.columns:

    if X_train[col].nunique() == 1:

        constant_cols.append(col)

print("Constant Features:", len(constant_cols))
```

```python
constant_cols
```

```python
X_train = X_train.drop(
    columns=constant_cols
)

X_test = X_test.drop(
    columns=constant_cols
)
```

```python
print(X_train.shape,X_test.shape)
```

```python
variance=X_train.var()
```

```python
variance.describe()
```

```python
variance.sort_values().head()
```

```python
from sklearn.feature_selection import VarianceThreshold

selector = VarianceThreshold(threshold=0.001)

selector.fit(X_train)
```

```python
selected_cols=X_train.columns[selector.get_support()]
X_train_low_var = X_train[selected_cols]
X_test_low_var = X_test[selected_cols]
```

```python
print(X_train.shape,X_train_low_var.shape)
```

```python
X_train = X_train_low_var.copy()
X_test = X_test_low_var.copy()
```

```python
corr=X_train.corr().abs()
```

```python
corr
```

```python
upper = corr.where(np.triu(np.ones(corr.shape),k=1).astype(bool))
```

```python
to_drop = []

for col in upper.columns:

    if any(upper[col] > 0.975):

        to_drop.append(col)
```

```python
len(to_drop)
```

```python
to_drop[:20]
```

```python
"Temperature (0 C)" in to_drop
```

```python
"Pressure (bar)" in to_drop
```

```python
X_train_corr = X_train.drop(
    columns=to_drop
)

X_test_corr = X_test.drop(
    columns=to_drop
)
```

```python
print(X_train_corr.shape)
print(X_test_corr.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(
    n_estimators=200,
    random_state=42,
    n_jobs=-1
)

rf.fit(X_train_corr, y_train)
```

```python
importance = pd.DataFrame({
    "Feature": X_train_corr.columns,
    "Importance": rf.feature_importances_
})

importance = importance.sort_values(
    by="Importance",
    ascending=False
)

importance.head(20)
```

```python
rf_top20 = set(importance.head(20)["Feature"])

```

```python
rf_features = importance["Feature"].head(20).tolist()
```

```python
rf_features = importance["Feature"].head(20).tolist()

X_rf_train = X_train_corr[rf_features]
X_rf_test = X_test_corr[rf_features]

print(X_rf_train.shape)
print(X_rf_test.shape)
```

```python
pd.DataFrame(
    rf_features,
    columns=["Descriptor"]
).to_excel("../data/RF_Top20_Features.xlsx",
    index=False
)
```

```python
X_rf_train.to_csv("../data/X_rf_train.csv", index=False)
X_rf_test.to_csv("../data/X_rf_test.csv", index=False)

y_train.to_csv("../data/y_train.csv", index=False)
y_test.to_csv("../data/y_test.csv", index=False)
```
