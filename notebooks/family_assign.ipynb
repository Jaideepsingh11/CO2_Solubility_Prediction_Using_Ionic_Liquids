```python
import pandas as pd 
import numpy as np 
import seaborn as sns
from rdkit import Chem
```

```python
df=pd.read_csv("../data/co2_solubility_cleaned.csv")
```

```python
df.sample(7)
```

```python
def family(name):
    name=str(name).lower()
    if 'imidazolium' in name:
        return "Imidazolium"
    if 'pyrrolidinium' in name:
        return "Pyrrolidinium"
    elif "pyridinium" in name:
        return "Pyridinium"
    elif "piperidinium" in name:
        return "Piperidinium"
    elif "phosphonium" in name:
        return "Phosphonium"
    elif "guanidinium" in name:
        return "Guanidinium"
    elif "sulfonium" in name:
        return "Sulfonium"
    elif "cyclopropenylium" in name:
        return "Cyclopropenylium"
    elif "ammonium" in name:
        return "Ammonium"
    elif "cholinium" in name:
        return "Ammonium"
    elif "aminium" in name:
        return "Ammonium"
    elif "triazabicyclo" in name:
        return "Amidinium"
    elif "hydrazinium" in name:
        return "Ammonium"
    elif "diamine" in name:
        return "Ammonium"
    elif "imidazol-3-ium" in name:
        return "Imidazolium"
    else:
        return "Others"
    

```

```python
family("1-butyl-3-methylimidazolium")
```

```python
df["Cation_Family"]=df["CATION NAME"].apply(family)
```

```python
df["Cation_Family"].value_counts()
```

```python
len(df[df["Cation_Family"]=="Imidazolium"])
```

```python
name=df["Cation_Family"].unique()
table=pd.DataFrame(name,columns=["Cation"])
table["T(Kelvin)"] = ""
table["P(bar)"] = ""
table["xCO2"] = ""
table["Data Points"]=""
table
```

```python
for cation in df["Cation_Family"].unique():
    table.loc[table["Cation"]==cation,"Data Points"]=str(len(df[df["Cation_Family"]==cation]))
    t_max=df[df["Cation_Family"]==cation]["Temperature (0 C)"].max()
    t_min=df[df["Cation_Family"]==cation]["Temperature (0 C)"].min()
    table.loc[table["Cation"]==cation,"T(Kelvin)"]=str(round(t_min,3))+"-"+str(round(t_max,3))
    p_max=df[df["Cation_Family"]==cation]["Pressure (bar)"].max()
    p_min=df[df["Cation_Family"]==cation]["Pressure (bar)"].min()
    table.loc[table["Cation"]==cation,"P(bar)"]=str(round(p_min,3))+"-"+str(round(p_max,3))
    x_max=df[df["Cation_Family"]==cation]["CO2 capacity (mol CO2/mol IL)"].max()
    x_min=df[df["Cation_Family"]==cation]["CO2 capacity (mol CO2/mol IL)"].min()
    table.loc[table["Cation"]==cation,"xCO2"]=str(round(x_min,3))+"-"+str(round(x_max,3))
```
